```python
print(skew(df['sqrt_Age']))
```

# K-means Clustering on Spending Score v/s Age of customers :

```python
from sklearn.cluster import KMeans
```

<h3>Selecting Features :</h3>

```python
v1 = df[['Age', 'Spending Score (1-100)']]
```

<h3>Using elbow method to choose k :</h3>

```python
inertia = []
range_k = range(1,10)
for k in range_k:
    k_means = KMeans(n_clusters=k , random_state=0)
    k_means.fit(v1)
    inertia.append(k_means.inertia_)

plt.plot(range_k, inertia)
plt.xlabel('Number of Clusters(k)')
plt.ylabel('Inertia')
plt.title('Choosing Optimal k by Elbow Method')
plt.show()
```

<h4>From above Elbow Curve, we start getting minimum value of inertia at k=4. Hence, k=4 can be chosen as an optimal value for clustering.</h4>

<h3>Running K-Means and Plotting CLusters:</h3>

```python
k = 4
k_means = KMeans(n_clusters = k , random_state = 0)
df['Cluster1'] = k_means.fit_predict(v1)
for cluster in range(k):
    cluster_data = df[df['Cluster1'] == cluster]
    plt.scatter(cluster_data['Age'] , cluster_data['Spending Score (1-100)'] , label = f"cluster{cluster}")
plt.xlabel('Age')
plt.ylabel('Spending Score (1-100)')
plt.title('K-Means Clustering for Spending Score v/s Age')
plt.show()
```

<h3>Examining The Cluster Centers :</h3>

```python
print(k_means.cluster_centers_)
```

# K-means Clustering on Spending Score v/s Annual Income :

```python
v2 = df[['Annual Income (k$)', 'Spending Score (1-100)']]
inertia = []
range_k = range(1,10)
for k in range_k:
    k_means = KMeans(n_clusters=k , random_state=0)
    k_means.fit(v2)
    inertia.append(k_means.inertia_)

plt.plot(range_k, inertia)
plt.xlabel('Number of Clusters(k)')
plt.ylabel('Inertia')
plt.title('Choosing Optimal k by Elbow Method')
plt.show()
```

<h3>From above Elbow Curve, we start getting minimum value of inertia at k=5. Hence, k=5 can be chosen as an optimal value for clustering.</h3>

```python
k = 5
k_means = KMeans(n_clusters = k , random_state = 0)
df['Cluster2'] = k_means.fit_predict(v2)
for cluster in range(k):
    cluster_data = df[df['Cluster2'] == cluster]
    plt.scatter(cluster_data['Annual Income (k$)'] , cluster_data['Spending Score (1-100)'] , label = f"cluster{cluster}")
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.title('K-Means Clustering for Spending Score v/s Annual Income')
plt.show()
```

<h3>Observing The Number of Customers in each Segment :</h3>

```python
labels = k_means.labels_
counts = np.bincount(labels)
for i, count in enumerate(counts):
    print(f"Cluster {i}: {count} data points")
```

```python
print(k_means.cluster_centers_)
```
