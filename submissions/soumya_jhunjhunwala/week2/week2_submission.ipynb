```python
import numpy as np
from sklearn.metrics import r2_score

def leaky_relu(x, alpha=0.01):
    return np.where(x > 0, x, alpha * x)

def leaky_relu_derivative(x, alpha=0.01):
    return np.where(x > 0, 1, alpha)

def mse_loss(y_true, y_pred):
    return np.mean((y_true - y_pred) ** 2)

class NeuralNetwork:
    def __init__(self, layers):
        self.layers = layers
        self.weights = []
        self.biases = []

        # Adam parameters
        self.m_weights = []
        self.v_weights = []
        self.m_biases = []
        self.v_biases = []
        self.t = 0 

        self._initialize_weights()

    def _initialize_weights(self):
        for i in range(len(self.layers) - 1):
            input_dim = self.layers[i]
            output_dim = self.layers[i + 1]
            weight = np.random.randn(input_dim, output_dim) * np.sqrt(2 / input_dim)
            bias = np.zeros((1, output_dim))
            self.weights.append(weight)
            self.biases.append(bias)

            self.m_weights.append(np.zeros_like(weight))
            self.v_weights.append(np.zeros_like(weight))
            self.m_biases.append(np.zeros_like(bias))
            self.v_biases.append(np.zeros_like(bias))

    def forward(self, X):
        self.z_values = []
        self.a_values = [X]

        for i in range(len(self.weights) - 1):
            z = np.dot(self.a_values[-1], self.weights[i]) + self.biases[i]
            a = leaky_relu(z)
            self.z_values.append(z)
            self.a_values.append(a)

        z = np.dot(self.a_values[-1], self.weights[-1]) + self.biases[-1]
        self.z_values.append(z)
        self.a_values.append(z)
        return z

    def backward(self, X, y, learning_rate=0.01, beta1=0.9, beta2=0.999, epsilon=1e-8):
        m = X.shape[0]
        self.t += 1

        grad_weights = [None] * len(self.weights)
        grad_biases = [None] * len(self.biases)

        dz = (self.a_values[-1] - y) / m
        for i in reversed(range(len(self.weights))):
            a_prev = self.a_values[i]
            dw = np.dot(a_prev.T, dz)
            db = np.sum(dz, axis=0, keepdims=True)
            grad_weights[i] = dw
            grad_biases[i] = db

            if i > 0:
                da_prev = np.dot(dz, self.weights[i].T)
                dz = da_prev * leaky_relu_derivative(self.z_values[i - 1])

        
        for i in range(len(self.weights)):
            
            self.m_weights[i] = beta1 * self.m_weights[i] + (1 - beta1) * grad_weights[i]
            self.v_weights[i] = beta2 * self.v_weights[i] + (1 - beta2) * (grad_weights[i] ** 2)

            self.m_biases[i] = beta1 * self.m_biases[i] + (1 - beta1) * grad_biases[i]
            self.v_biases[i] = beta2 * self.v_biases[i] + (1 - beta2) * (grad_biases[i] ** 2)

            m_hat_w = self.m_weights[i] / (1 - beta1 ** self.t)
            v_hat_w = self.v_weights[i] / (1 - beta2 ** self.t)
            m_hat_b = self.m_biases[i] / (1 - beta1 ** self.t)
            v_hat_b = self.v_biases[i] / (1 - beta2 ** self.t)

            self.weights[i] -= learning_rate * m_hat_w / (np.sqrt(v_hat_w) + epsilon)
            self.biases[i] -= learning_rate * m_hat_b / (np.sqrt(v_hat_b) + epsilon)

    def train(self, X, y, epochs=1000, learning_rate=0.01, verbose=True):
        for epoch in range(epochs):
            y_pred = self.forward(X)
            loss = mse_loss(y, y_pred)
            self.backward(X, y, learning_rate)
            if verbose and epoch % 100 == 0:
                print(f"Epoch {epoch} - Loss: {loss:.4f}")

    def predict(self, X):
        return self.forward(X)
```

```python
from sklearn.metrics import r2_score

def evaluate_regression_metrics(y_true, y_pred):
    mse = np.mean((y_true - y_pred) ** 2)
    rmse = np.sqrt(mse)
    mae = np.mean(np.abs(y_true - y_pred))
    r2 = r2_score(y_true, y_pred)

    print(f"MSE:  {mse:.4f}")
    print(f"RMSE: {rmse:.4f}")
    print(f"MAE:  {mae:.4f}")
    print(f"R²:   {r2:.4f}")
```

```python
model = NeuralNetwork(layers=[X_train.shape[1], 128, 64, 1])
model.train(X_train, y_train, epochs=1000, learning_rate=0.001, verbose=True)

y_pred = model.predict(X_test)
evaluate_regression_metrics(y_test.reshape(-1,1), y_pred)
```
