Log Transformation

```python
data['log age']=np.log(data['Age'])
data['log annual income']=np.log(data['Annual Income'])

plt.title("Age after using log transform")
plt.hist(data['log age'], bins=10,edgecolor='black')
plt.xlabel('Frequency')
plt.ylabel('Age')
plt.show()
plt.title("Annual Income after using log transform")
plt.hist(data['log annual income'], bins=10,edgecolor='black')
plt.xlabel('Frequency')
plt.ylabel('Annual Income')
plt.show()

mean_income=data['Annual Income'].mean()
median_income=data['Annual Income'].median()
sd_income=data['Annual Income'].std()
skew_income=(3*(mean_income-median_income))/sd_income
mean_age=data['Age'].mean()
median_age=data['Age'].median()
sd_age=data['Age'].std()
skew_age=(3*(mean_age-median_age))/sd_age

mean_income_log=data['log annual income'].mean()
median_income_log=data['log annual income'].median()
sd_income_log=data['log annual income'].std()
skew_income_log=(3*(mean_income_log-median_income_log))/sd_income_log
mean_age_log=data['log age'].mean()
median_age_log=data['log age'].median()
sd_age_log=data['log age'].std()
skew_age_log=(3*(mean_age_log-median_age_log))/sd_age_log

print(f'Initial Skewness for income : {skew_income} After log transform : {skew_income_log}')
print(f'Initial mean for income : {mean_income} After log transform : {mean_income_log}')
print(f'Initial Skewness for age : {skew_age} After log age : {skew_age_log}')
print(f'Initial mean for age : {mean_age} After log age : {mean_age_log}')
```

StandardScaler

```python
scaler = StandardScaler()
data_include = data.select_dtypes(include=['number'])
scaled = scaler.fit_transform(data_include)
data_scaled = pd.DataFrame(scaled, columns=data_include.columns)

plt.hist(data_scaled['Age'], bins=10,edgecolor='black')
plt.ylabel('Age')
plt.show()
plt.hist(data_scaled['Annual Income'], bins=10,edgecolor='black')
plt.ylabel('Annual Income')
plt.show()

mean_income_scaled = data_scaled['Annual Income'].mean()
median_income_scaled = data_scaled['Annual Income'].median()
sd_income_scaled = data_scaled['Annual Income'].std()
skew_income_scaled = (3 * (mean_income_scaled - median_income_scaled)) / sd_income_scaled

mean_age_scaled = data_scaled['Age'].mean()
median_age_scaled = data_scaled['Age'].median()
sd_age_scaled = data_scaled['Age'].std()
skew_age_scaled = (3 * (mean_age_scaled - median_age_scaled)) / sd_age_scaled

# Print results
print(f'Skewness for income after Standard Scaling : {skew_income_scaled}')
print(f'Mean for income after Standard Scaling     : {mean_income_scaled}')
print(f'Skewness for age after Standard Scaling    : {skew_age_scaled}')
print(f'Mean for age after Standard Scaling        : {mean_age_scaled}')
```

MinMaxScaler

```python
scaler = MinMaxScaler()
data_include = data.select_dtypes(include=['number'])
scaled = scaler.fit_transform(data_include)
data_scaled = pd.DataFrame(scaled, columns=data_include.columns)

plt.hist(data_scaled['Age'], bins=10,edgecolor='black')
plt.ylabel('Age')
plt.show()
plt.hist(data_scaled['Annual Income'], bins=10,edgecolor='black')
plt.ylabel('Annual Income')
plt.show()

mean_income_scaled = data_scaled['Annual Income'].mean()
median_income_scaled = data_scaled['Annual Income'].median()
sd_income_scaled = data_scaled['Annual Income'].std()
skew_income_scaled = (3 * (mean_income_scaled - median_income_scaled)) / sd_income_scaled

mean_age_scaled = data_scaled['Age'].mean()
median_age_scaled = data_scaled['Age'].median()
sd_age_scaled = data_scaled['Age'].std()
skew_age_scaled = (3 * (mean_age_scaled - median_age_scaled)) / sd_age_scaled

print(f'Skewness for income after  MinMax scaling : {skew_income_scaled:.4f}')
print(f'Mean for income after  MinMax scaling     : {mean_income_scaled:.2f}\n')
print(f'Skewness for age after  MinMax scaling    : {skew_age_scaled:.4f}')
print(f'Mean for age after  MinMax scaling        : {mean_age_scaled:.2f}')
```

Clustering with K-Means

```python
kmeans = KMeans(n_clusters=3)
data['cluster']=kmeans.fit_predict(data[['Annual Income','Spending Score']])

df1=data[data['cluster']==0]
df2=data[data['cluster']==1]
df3=data[data['cluster']==2]

plt.xlabel('Annual Income')
plt.ylabel('Spending Score')
plt.scatter(x=df1['Annual Income'],y=df1['Spending Score'],color='black')
plt.scatter(x=df2['Annual Income'],y=df2['Spending Score'],color='blue')
plt.scatter(x=df3['Annual Income'],y=df3['Spending Score'],color='green')
plt.show()

#Judging from the scatter plot df1 = high income low spenders, df2 = high income high spenders , df3 = low income average spenders
```
