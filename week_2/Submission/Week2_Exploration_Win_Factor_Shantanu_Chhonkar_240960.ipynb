# Data Exploration with Pandas


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import r2_score,mean_squared_error
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('/content/mw_pw_profiles.txt')
df.to_csv('mw_pw_profiles.csv')
print(df.head())
```

```python
df.info()
```

```python
numeric_df = df.select_dtypes(include=[np.number])
```

```python
plt.figure(figsize=(40,40))
sns.heatmap(numeric_df.corr(),annot=True,cmap='coolwarm')
plt.show()
```

```python
plt.figure(figsize=(10,8))
plt.scatter(df['runs_scored'],df['fantasy_score_batting'],color='blue')
plt.title('Runs scored vs Fantasy score batting')
plt.xlabel('runs_scored')
plt.ylabel('fantasy_score_batting')
plt.show()
```
