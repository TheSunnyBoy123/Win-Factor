# Adding New Features
<h3> 1. Calculating the impact of the player as batsman</h3>
Using runs scored (1 point for each run scored) , fours scored(4 extra points for each four) , sixes scored(6 extra points for each six) , we will calculate the 'bat_impact_score' of each player in each match.

```python
df['bat_impact_score'] = df['runs_scored'] + 4 * df['fours_scored'] + 6 * df['sixes_scored']
```

```python
plt.scatter(df['bat_impact_score'] , df['fantasy_score_batting'] , s=20 , edgecolor = 'Black' , color = 'Pink')
plt.xlabel('Batting Impact Score')
plt.ylabel('Fantasy Score(Batting)')
plt.title('Scatter Plot : Fantasy Score(Batting) v/s Impact Score(Batting) calculated')
plt.grid()
plt.show()
```

<h3> 2. Calculating the impact of the player as bowler</h3>
Using wickets taken (25 points for each wicket taken), lbw done (5 extra points), bowled done (5 extra points) , maiden overs bowled(8 extra points for each maiden over) , dot balls(1 points for each dot ball), we will calculate the 'bowl_impact_score' of each player in each match.

```python
df['bowl_impact_score'] = 1.5*df['dot_balls_as_bowler'] + 25 * df['wickets_taken'] + 8 * df['maidens'] +5*df['bowled_done'] + 5*df['lbw_done']- 0.3 * df['runs_conceded']
```

```python
plt.scatter(df['bowl_impact_score'] , df['fantasy_score_bowling'] , s=20 , edgecolor = 'Black' , color = 'Pink')
plt.xlabel('Bowling Impact Score')
plt.ylabel('Fantasy Score(Bowling)')
plt.title('Scatter Plot : Fantasy Score(Bowling) v/s Impact Score(Bowling) calculated')
plt.grid()
plt.show()
```

<h3> 3. Calculating the impact of the player as fielder</h3>
Using catches taken (8 points for each catch taken) , direct-hit run outs done(12 points for each direct-hit run out), run-out throw(6 points), stumpings done (12 points for each stump out), we will calculate the 'field_impact_score' of each player in each match.

```python
df['field_impact_score'] = 8*df['catches_taken'] + 12*df['run_out_direct'] + 6*df['run_out_throw'] + 12*df['stumpings_done']
```

<h3> 4. Calculating the total impact of the player in the match</h3>
Total Impact = Impact as Batsman + Impact as Bowler + Impact as Fielder

```python
df['total_impact'] = df['bat_impact_score'] + df['bowl_impact_score'] + df['field_impact_score']
```

```python
plt.scatter(df['total_impact'] , df['fantasy_score_total'] , s=20 , edgecolor = 'Black' , color = 'Pink')
plt.xlabel('Total Impact Score')
plt.ylabel('Total Fantasy Score')
plt.title('Scatter Plot : Total Fantasy Score v/s Impact Score(Total) calculated')
plt.grid()
plt.show()
```
